```python
#le niveau de stress des etudiants
data['Stress Level '].value_counts()
```

```python
stress_level=list(data['Stress Level '].value_counts().keys())
count_stress_level=list(data['Stress Level '].value_counts().values)
```

```python
plt.figure(figsize=(20,10))
explode=[0.1,0,0,0]
palete_color=sns.color_palette('dark') #coleur du graphes

#faire le graphe des donnee
plt.pie(count_stress_level,labels=stress_level,colors=palete_color,explode=explode,autopct='%.0f%%')
```

```python
data['Financial Status'].value_counts()
```

```python
f_status=list(data['Financial Status'].value_counts().keys())
f_statut_count=list(data['Financial Status'].value_counts().values)
```

```python
plt.figure(figsize=(20,10))
explode=[0.10,0,0,0]
palete_color=sns.color_palette('dark') #coleur du graphes

#faire le graphe des donnee
plt.pie(f_statut_count,labels=f_status,colors=palete_color,explode=explode,autopct='%.0f%%')
```
